```python
import pandas as pd
```

```python
bank_data = pd.read_csv('bank.csv', sep=';')
```

```python
bank_data.head(10)
```

```python
bank_data.info()
```

```python
pd.get_dummies(bank_data.marital, prefix='marital')

marital_dummies = pd.get_dummies(bank_data.marital, prefix='marital').ix[:, 1:]

bank_data = pd.concat([bank_data, marital_dummies], axis=1)
```

```python
pd.get_dummies(bank_data.education, prefix='education')

education_dummies = pd.get_dummies(bank_data.education, prefix='education').ix[:, 1:]

bank_data = pd.concat([bank_data, education_dummies], axis=1)
```

```python
pd.get_dummies(bank_data.default, prefix='default')

default_dummies = pd.get_dummies(bank_data.default, prefix='default').ix[:, 1:]

bank_data = pd.concat([bank_data, default_dummies], axis=1)
```

```python
pd.get_dummies(bank_data.housing, prefix='housing')

housing_dummies = pd.get_dummies(bank_data.housing, prefix='housing').ix[:, 1:]

bank_data = pd.concat([bank_data, housing_dummies], axis=1)
```

```python
pd.get_dummies(bank_data.loan, prefix='loan')

loan_dummies = pd.get_dummies(bank_data.loan, prefix='loan').ix[:, 1:]

bank_data = pd.concat([bank_data, loan_dummies], axis=1)
```

```python
bank_data.head()
```

```python
pd.get_dummies(bank_data.job, prefix='job').head(10)
```

```python
job_dummies = pd.get_dummies(bank_data.job, prefix='job').ix[:, 1:]

bank_data = pd.concat([bank_data, job_dummies], axis=1)

bank_data.head()
```

```python
pd.get_dummies(bank_data.contact, prefix='contact').head(10)
```

```python
contact_dummies = pd.get_dummies(bank_data.contact, prefix='contact').ix[:, 1:]

bank_data = pd.concat([bank_data, contact_dummies], axis=1)
```

```python
pd.get_dummies(bank_data.poutcome, prefix='poutcome').head(10)
```

```python
poutcome_dummies = pd.get_dummies(bank_data.poutcome, prefix='poutcome').ix[:, 1:]

bank_data = pd.concat([bank_data, poutcome_dummies], axis=1)
```

```python
bank_data['y'] = bank_data.y.map({'no':0, 'yes':1})
```

```python
bank_data.head()
```

```python
bank_data.info()
```

```python
bank_data['y'] = bank_data['y'].replace(to_replace=[1], value=1.0)

features = bank_data.ix[:, 17:].values
target = bank_data.y.values

print type(target)
print type(bank_data.y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import cross_val_score
model = DecisionTreeClassifier(random_state=0)

scores = cross_val_score(model, features, target, cv=5,scoring = 'accuracy')
```

```python
import numpy as np
np.mean(scores)
```

```python
model.fit(features, target)
```

```python
model.feature_importances_
```

```python
feature_columns = ['marital_married','marital_single','education_secondary','education_tertiary','education_unknown','default_yes','housing_yes','loan_yes','job_blue-collar','job_entrepreneur','job_housemaid','job_management','job_retired','job_self-employed','job_services','job_student','job_technician','job_unemployed','job_unknown','contact_telephone','contact_unknown','poutcome_other','poutcome_success','poutcome_unknown']

pd.DataFrame({'feature':feature_columns, 'importance':model.feature_importances_}).sort('importance',ascending=False)
```
