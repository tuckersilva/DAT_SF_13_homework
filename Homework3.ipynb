#### Question 2

```python
categorical_variables = ['job','marital','education','contact','month','poutcome']
binary_variables = ['default', 'housing','loan','y']

string_variables = categorical_variables + binary_variables
```

```python
from sklearn import preprocessing

bank_numerical = bank_data.copy()
encoder_dict = {}

for col in string_variables:
    le = preprocessing.LabelEncoder()
    le.fit(bank_data[col])
    encoder_dict[col] = le
    bank_numerical[col] = le.transform(bank_data[col])
```

```python
bank_numerical.head()
```

```python
encoder_dict
```

```python
encoder_dict['poutcome'].classes_
```

#### Question 3

```python
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
xcols = [col for col in bank_numerical.columns if col != 'y']
```

```python
X = bank_numerical[xcols].values
y = bank_numerical['y'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

```

```python
clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

conf_mat = confusion_matrix(y_test,y_pred)

print conf_mat
```

#### Question 4

```python
from sklearn import cross_validation
scores = cross_validation.cross_val_score(clf, X, y, cv=cross_validation.KFold(len(X), n_folds=5, shuffle=True, random_state=0))

scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() *2))
```
